```python
import matplotlib.pyplot as plt
import pandas as pd
import csv


with open('amazon_jobs_dataset.csv', encoding ='UTF-8') as file_obj:
    file_data = csv.DictReader(file_obj, skipinitialspace=True)

    dct = {}
    for row in file_data:
        date = row['Posting_date'].split()
        key = date[2]
        if key in dct:
            dct[key] += 1
        else:
            dct[key] = 1
    sorted_dict = dict(sorted(dct.items()))
    x=sorted_dict.keys()
    y=sorted_dict.values()
    plt.plot(x,y)
    plt.xlabel("Year")
    plt.ylabel("Jobs")
    plt.title('Year vs Job_Openings')
    plt.show()
    
    for i in sorted_dict.items():
        print(i[0],i[1])
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import csv


with open('amazon_jobs_dataset.csv', encoding ='UTF-8') as file_obj:
    file_data = csv.DictReader(file_obj, skipinitialspace=True)

    dct = {}
    for row in file_data:
        date = row['Posting_date'].split()
        key = date[0]
        if key in dct:
            dct[key] += 1
        else:
            dct[key] = 1
    sorted_dict = dict(sorted(dct.items()))
    x=sorted_dict.keys()
    y=sorted_dict.values()
    plt.bar(x,y)
    plt.xlabel("Month")
    plt.ylabel("Jobs")
    plt.title('Month vs Job_Openings')
    plt.show()
    
    for i in sorted_dict.items():
        print(i[0],i[1])
```
