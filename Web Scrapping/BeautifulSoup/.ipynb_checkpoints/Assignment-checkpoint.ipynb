```python
import requests
from bs4 import BeautifulSoup

base_url = 'https://www.imdb.com/search/title/?release_date={}-01-01,{}-12-31&sort=num_votes,desc&ref_=adv_prv'
names=[]
years = range(2010, 2015)
for year in years:
    url = base_url.format(year, year)
    response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'})
    html_data = response.content
    data = BeautifulSoup(html_data, 'html.parser')
    movies = data.find_all(class_="ipc-title ipc-title--base ipc-title--title ipc-title-link-no-icon ipc-title--on-textPrimary sc-b189961a-9 iALATN dli-title")
    for movie in movies:
        name = movie.find(class_="ipc-title__text").string
        names.append(name)
        stripped_names = [name.strip('1234567890. ') for name in names]

# Print the stripped names
for name in stripped_names:
    print(name)
```

```python
from bs4 import BeautifulSoup
import requests
import time
from random import randint

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}

dct = {}

# Loop through multiple pages (1 to 5) of IMDb search results
for page in range(1, 6):
    res = requests.get(f'https://www.imdb.com/search/title/?release_date=2018&sort=num_votes,desc&page={page}&ref_=adv_nxt', headers=headers)
    data = BeautifulSoup(res.text, "html.parser")
    tags = data.find_all('div', class_='lister-item')

    # Extract movie names and durations from each page
    for tag in tags:
        if tag.find('span', class_='runtime'):
            head = tag.find('h3', class_='lister-item-header')
            dur = tag.find('span', class_='runtime')
            duration = int(dur.text.strip().split(' ')[0])
            dct[head.a.string] = duration
    time.sleep(randint(0, 3))

# Find the movie with the maximum duration
max_duration = max(dct.values())
max_movies = [movie for movie, duration in dct.items() if duration == max_duration]

# Print the movie(s) with the maximum duration
for movie in max_movies:
    print(movie, max_duration)
```

```python
from bs4 import BeautifulSoup
import requests
import time
from random import randint

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}

def minutes(duration_string):
    if "h" and "m" in duration_string:
    # Split the string into hours and minutes
        hours, minutes = duration_string.split('h ')
        minutes = minutes.strip('m')
    else:
        return None
    # Convert hours and minutes to integers
    hours = int(hours)
    minutes = int(minutes)

    # Calculate the total duration in minutes
    total_minutes = hours * 60 + minutes
    return total_minutes
dct = {}
for i in range(1, 202, 50):  # Adjusted loop range
    url = f'https://www.imdb.com/search/title/?release_date=2018&sort=num_votes,desc&page={i}&ref_=adv_nxt'
    res = requests.get(url, headers=headers)  # Fixed URL generation
    data = BeautifulSoup(res.text, "html.parser")
    tags = data.find_all('div', class_="sc-ab6fa25a-3 bVYfLY dli-parent")
    for j in tags:
        head = j.find('h3', class_='ipc-title__text')
        dur = j.find_all('span', class_='sc-b189961a-8 kLaxqf dli-title-metadata-item')[1].string
        if dur:  # Check if duration is available
            t = minutes(dur)
            dct[head.string] = t
    time.sleep(randint(0, 3))
dct = {key: value for key, value in dct.items() if value is not None}
# Find the title with the maximum duration
max_dur = max(dct.values(), default=-1)
max_name = max((title, duration) for title, duration in dct.items() if duration == max_dur)

print(max_name[0], max_name[1])
```
