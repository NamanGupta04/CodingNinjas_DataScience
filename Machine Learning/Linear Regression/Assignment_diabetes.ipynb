```python
import csv
import numpy as np
```

```python
data = np.loadtxt("training_diabetes_x_y_train.csv", delimiter=",")
data2 = np.loadtxt("test_diabetes_x_test.csv" , delimiter = ",")
data.shape
```

```python
x_train = data[:,:10]
y_train = data[:,10:11]

x_test = data2[:]

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
alg1 = LinearRegression()
alg1.fit(x_train, y_train)
```

```python
y_pred = alg1.predict(x_test)
print(alg1.coef_)
print(alg1.intercept_)

y_pred = np.round(y_pred,decimals = 5)

data3  = np.savetxt("predictions.csv" , y_pred ,delimiter =',')
a = open("predictions.csv" , "r")
a.read()
```
